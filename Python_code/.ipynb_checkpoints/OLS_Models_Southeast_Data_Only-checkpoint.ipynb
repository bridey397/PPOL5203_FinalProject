**Life Expectancy and Risk Score**

```python
se_model_ols_r_bivariate = smf.ols(formula='Life_Expectancy ~ RISK_SCORE', data=Southeast_county_data).fit()

print(se_model_ols_r_bivariate.summary())
```

```python
Southeast_county_data['RISK_SCORE_SQ'] = Southeast_county_data['RISK_SCORE']**2 
```

```python
se_model_ols_r_sq = smf.ols(formula='Life_Expectancy ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density', data=Southeast_county_data).fit()

print(se_model_ols_r_sq.summary())
```

```python
se_model_ols_r_sq2 = smf.ols(formula='Life_Expectancy ~ RISK_SCORE + RISK_SCORE_SQ + Community_Type + Population_Density', data=Southeast_county_data).fit()

print(se_model_ols_r_sq2.summary())
```

```python
#no squared term but interaction between community and risk score
se_model_ols_r_int = smf.ols(formula='Life_Expectancy ~ RISK_SCORE + Population_Density + Community_Type + RISK_SCORE*Community_Type', data=Southeast_county_data).fit()

print(se_model_ols_r_int.summary())
```

```python
sns.lmplot(x="RISK_SCORE", y="Life_Expectancy", data=Southeast_county_data,
           order=2, ci=None, scatter_kws={"s": 20})

plt.show()
```

```python
sns.lmplot(data=Southeast_county_data, x="RISK_SCORE", y="Life_Expectancy", 
           hue = "Community_Type", markers = ["o", "x", "+"],
           palette="Set1",
        ci=None, scatter_kws={"s": 20})

plt.show()
```

```python
#quadratic model control for community type
sns.lmplot(data=Southeast_county_data, x="RISK_SCORE", y="Life_Expectancy", 
           order=2,
           hue = "Community_Type", markers = ["o", "x", "+"],
           palette="Set1",
        ci=None, scatter_kws={"s": 20})

plt.show()
```

```python
se_model_ols_r_lifeexp_sq3 = smf.ols(formula='Life_Expectancy ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density + Community_Type + RISK_SCORE*Community_Type', data=Southeast_county_data).fit()

print(se_model_ols_r_lifeexp_sq3.summary())
```

```python
anova_results = anova_lm(se_model_ols_r_sq3, se_model_ols_r_sq)
print(anova_results)
```

**Age-adjusted Death Rate and Risk Score**

```python
se_model_ols_r_sq = smf.ols(formula='Age_Adjusted_Death_Rate ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density', data=Southeast_county_data).fit()

print(se_model_ols_r_sq.summary())
```

```python
se_model_ols_r_sq2 = smf.ols(formula='Age_Adjusted_Death_Rate ~ RISK_SCORE + RISK_SCORE_SQ + Community_Type + Population_Density', data=Southeast_county_data).fit()

print(se_model_ols_r_sq2.summary())
```

```python
sns.lmplot(x="RISK_SCORE", y="Age_Adjusted_Death_Rate", data=Southeast_county_data,
           order=2, ci=None, scatter_kws={"s": 20})

plt.show()
```

```python
sns.lmplot(data=Southeast_county_data, x="RISK_SCORE", y="Age_Adjusted_Death_Rate", 
           hue = "Community_Type", markers = ["o", "x", "*"],
           palette="Set1",
        ci=None, scatter_kws={"s": 20})

plt.show()
```

```python
#quadratic model control for community type
sns.lmplot(data=Southeast_county_data, x="RISK_SCORE", y="Age_Adjusted_Death_Rate", 
           order=2,
           hue = "Community_Type", markers = ["o", "x", "*"],
           palette="Set1",
        ci=None, scatter_kws={"s": 20})

plt.show()
```

```python
se_model_ols_r_dr_int = smf.ols(formula='Age_Adjusted_Death_Rate ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density + Community_Type + RISK_SCORE*Community_Type', data=Southeast_county_data).fit()

print(se_model_ols_r_dr_int.summary())
```

```python
se_model_ols_r_nosq = smf.ols(formula='Age_Adjusted_Death_Rate ~ RISK_SCORE + Population_Density + Community_Type + RISK_SCORE*Community_Type', data=Southeast_county_data).fit()

print(se_model_ols_r_nosq.summary())
```

```python
beta1pop = se_model_ols_r_int.params[-1]
```

```python
SE = Southeast_county_data.copy()

SE['age_pop'] = SE['Age_Adjusted_Death_Rate'] - (beta1pop*(SE['Population_Density'].mean()))

print(SE['age_pop'].head())
print(SE['Age_Adjusted_Death_Rate'].head())
beta1pop * SE['Population_Density'].mean()

SE.head()
```

```python
(ggplot(SE) + aes(x = 'RISK_SCORE', y = 'age_pop', fill = "Community_Type") + geom_point())
```

```python
#interactions don't appear stat sig based on anova results comparing the models 
anova_results_sq_with_commtype = anova_lm(se_model_ols_r_sq, se_model_ols_r_sq2)
anova_results_sq_with_inter = anova_lm(se_model_ols_r_sq2, se_model_ols_r_int)
anova_results_inter_with_nosq = anova_lm(se_model_ols_r_int, se_model_ols_r_nosq)

print(anova_results_sq_with_commtype)
print(anova_results_sq_with_inter)
print(anova_results_inter_with_nosq)
```

**Physically Unhealthy and Risk Score**

```python
se_phys_model_ols_r_sq = smf.ols(formula='Average_Number_of_Physically_Unhealthy_Days ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density', data=Southeast_county_data).fit()

print(se_phys_model_ols_r_sq.summary())
```

```python
se_phys_model_ols_r = smf.ols(formula='Average_Number_of_Physically_Unhealthy_Days ~ RISK_SCORE + Community_Type + Population_Density', data=Southeast_county_data).fit()

print(se_phys_model_ols_r.summary())
```

```python
se_phys_model_ols_r_int = smf.ols(formula='Average_Number_of_Physically_Unhealthy_Days ~ RISK_SCORE + Population_Density + Community_Type + RISK_SCORE*Community_Type', data=Southeast_county_data).fit()

print(se_phys_model_ols_r_int.summary())
```

```python
beta1pop = phys_model_ols_r_int.params[-1]

SE['phys_pop'] = SE['Average_Number_of_Physically_Unhealthy_Days'] - (beta1pop*(SE['Population_Density'].mean()))

SE.head()
```

```python
(ggplot(SE) + aes(x = 'RISK_SCORE', y = 'phys_pop', fill = "Community_Type") + geom_point())

```

```python
se_phys_model_ols_r_sq3 = smf.ols(formula='Average_Number_of_Physically_Unhealthy_Days ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density + Community_Type + RISK_SCORE*Community_Type', data=Southeast_county_data).fit()

print(se_phys_model_ols_r_sq3.summary())
```

```python
#squared term doesn't appear statistically significant
anova_results_sq_with_commtype = anova_lm(se_phys_model_ols_r_sq, se_phys_model_ols_r_sq2)
anova_results_sq_with_inter = anova_lm(se_phys_model_ols_r_sq2, se_phys_model_ols_r_sq3)
anova_results_inter_with_nosq = anova_lm(se_phys_model_ols_r_int, se_phys_model_ols_r_sq3)

print(anova_results_sq_with_commtype)
print(anova_results_sq_with_inter)
print(anova_results_inter_with_nosq)
```

**Mentally Unhealthy Days**

```python
se_ment_model_ols_r = smf.ols(formula='Average_Number_of_Mentally_Unhealthy_Days ~ RISK_SCORE + Population_Density + Community_Type', data=Southeast_county_data).fit()

print(se_ment_model_ols_r.summary())
```

```python
se_ment_model_ols_r_int = smf.ols(formula='Average_Number_of_Mentally_Unhealthy_Days ~ RISK_SCORE + Population_Density + Community_Type + Community_Type*RISK_SCORE', data=Southeast_county_data).fit()

print(se_ment_model_ols_r_int.summary())
```

```python
se_ment_model_ols_r_sq = smf.ols(formula='Average_Number_of_Mentally_Unhealthy_Days ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density', data=Southeast_county_data).fit()

print(se_ment_model_ols_r_sq.summary())
```

```python
se_ment_model_ols_r_sq2 = smf.ols(formula='Average_Number_of_Mentally_Unhealthy_Days ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density + Community_Type', data=Southeast_county_data).fit()

print(se_ment_model_ols_r_sq2.summary())
```

```python
se_ment_model_ols_r_sq_int = smf.ols(formula='Average_Number_of_Mentally_Unhealthy_Days ~ RISK_SCORE + RISK_SCORE_SQ + Population_Density + Community_Type + RISK_SCORE*Community_Type', data=Southeast_county_data).fit()

print(se_ment_model_ols_r_sq_int.summary())
```

```python
sns.lmplot(x="RISK_SCORE", y="Average_Number_of_Mentally_Unhealthy_Days", data=Southeast_county_data,
           order=2, ci=None, scatter_kws={"s": 20})

plt.show()
```

```python
#squared term doesn't appear statistically significant
anova_results_sq_with_nocommtype = anova_lm(se_ment_model_ols_r_sq, se_ment_model_ols_r_sq2)
anova_results_sq_with_nointer = anova_lm(se_ment_model_ols_r_sq2, se_phys_model_ols_r_int)
anova_results_inter_with_nosq = anova_lm(se_ment_model_ols_r_int, se_ment_model_ols_r_sq_int)

print(anova_results_sq_with_nocommtype)
print(anova_results_sq_with_nointer)
print(anova_results_inter_with_nosq)
```

```python
models = [se_model_ols_r_lifeexp_sq3, se_model_ols_r_dr_int, se_phys_model_ols_r_int, se_ment_model_ols_r_sq_int]

stargazertable = Stargazer(models)

stargazertable.title("Regression Output for Dependent Health Outcomes")

stargazertable

stargazertable.custom_columns(['Life Expectancy', 'Age-Adjusted Death Rate', "Physically Unhealthy Days", "Mentally Unhealthy Days"], [1, 1, 1, 1])

stargazertable.show_model_numbers(False)

stargazertable.significant_digits(4)

stargazertable.show_degrees_of_freedom(False)

stargazertable.covariate_order(['RISK_SCORE', 'RISK_SCORE_SQ', 'Community_Type[T.Small Urban]', "Community_Type[T.Urban]", 'RISK_SCORE:Community_Type[T.Small Urban]','RISK_SCORE:Community_Type[T.Urban]', "Population_Density"])

stargazertable

HTML(stargazertable.render_html())
```
